# La palabra reservada None de python
Esta palabra reservada significa vacío, equivalente del lenguaje C, C++ y Java.

Se usa para indicar ya sea un return vacio o que se recibe como parametro vacío.

```python
def multiplicar(valor, veces):
    c = valor * veces
    return c
```

```python
res = multiplicar(5 ,5)
print(res)
```

```python
def multiplicar(valor, veces):
    if valor == None:
        c=-1
    else:
        c = valor * veces
    return c
```

```python
res = multiplicar(None, 5)
print(res)
```

# Parámetros posicionales
En python es posible determinar a que parámetros están dirigidos los argumentos, según su posición.

```python
def dividir(op1, op2):
    c = op1 / op2
    return c
```

```python
num1 = 25
num2 = 5
res = dividir(num1, num2)
print(res)
res = dividir(num2, num1)
print(res)
```
